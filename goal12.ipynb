```python
SYSTEM_PROMPT = """
You are an assistant for SJSU Dining Services helping reduce food waste on campus.

Your job is to write a short, friendly pickup alert for SJSU students when surplus food
is available at the end of a meal service. The alert will be posted to the SAMMY App
so students can come pick up free food before it is discarded.

Guidelines:
- Write exactly 3–4 sentences.
- Use a warm, welcoming tone — not clinical or bureaucratic.
- Include: what food is available, how much, where to pick it up, and by when.
- End with a short encouraging line (e.g., "Don't let good food go to waste!").
- Do NOT invent details not provided by the staff member.
- If the input includes a language instruction, write the alert in that language.
"""

def generate_alert(staff_input, system_prompt=SYSTEM_PROMPT):
    full_prompt = f"{system_prompt}\n\nStaff input:\n{staff_input}"
    response = model.generate_content(full_prompt)
    return response.text

print("✅ System prompt and generate_alert() function defined.")
```

```python
def validated_alert_generator(food_type, quantity, pickup_time, location, food_condition):
    required_fields = {
        "food_type": food_type,
        "quantity": quantity,
        "pickup_time": pickup_time,
        "location": location
    }

    vague_keywords = ["unknown", "not sure", "maybe", "sometime", "somewhere", "unclear", ""]

    for field_name, value in required_fields.items():
        if not value or any(keyword in value.lower() for keyword in vague_keywords):
            return (
                f"🚫 ALERT BLOCKED — Input validation failed.\n"
                f"Field '{field_name}' is missing or too vague: '{value}'\n"
                f"Please fill in all fields with specific information.\n"
                f"If you are unsure, contact a supervisor before proceeding."
            )

    food_condition = food_condition.lower().strip()

    if food_condition in ["unsure", "do not donate", "unsafe"]:
        return (
            f"🚫 ALERT BLOCKED — Food safety check failed.\n"
            f"Food condition marked as: '{food_condition}'\n"
            f"No alert can be posted until a supervisor confirms the food is safe.\n"
            f"Please contact your dining supervisor immediately."
        )

    if food_condition != "safe":
        return (
            f"🚫 ALERT BLOCKED — Unrecognized food condition: '{food_condition}'\n"
            f"Please enter one of: 'safe', 'unsure', or 'do not donate'."
        )

    staff_input = f"""
Food type: {food_type}
Quantity: {quantity}
Pickup window: {pickup_time}
Location: {location}
"""
    alert = generate_alert(staff_input)
    return f"✅ VALIDATION PASSED — Alert generated:\n\n{alert}"

print("✅ validated_alert_generator() function defined.")
```

```python
print("🧪 VALIDATION TEST A — Vague/unsafe input (should be BLOCKED)")
print("=" * 60)

result_a = validated_alert_generator(
    food_type="Leftover event food, mixed items",
    quantity="Unknown",
    pickup_time="Sometime after 6",
    location="Somewhere near the event hall",
    food_condition="unsure"
)

print(result_a)
```
